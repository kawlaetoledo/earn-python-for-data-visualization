1. Importação da base de dados:


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
dados = pd.read_csv(r'C:\Users\kawla\Documents\curso\pandas\athlete_events.csv')
```

2. Mostrando os dados em tabela:

```python
dados.head()
```

3. Separando os dados do sexo masculino e mostrando na tela: 

```python
masculinos = dados.loc[dados['Sex']=='M']
```

```python
masculinos.head()
```

4. Criando as duas váriaveis que precisamos para criação do gráfico de dispersão, uma sendo para áltura e outro para o peso:

```python
a = masculinos ['Height']
p = masculinos ['Weight']
```

5. Criando e chamando o gráfico para aparecer na tela e analisar relações entre duas variáveis. 

```python
plt.scatter(a,p)
plt.show()
```

Os gráficos de dispersão ou Scatter plot são representações gráficas do relacionamento entre duas variáveis numéricas. O Scatter plot utiliza pontos para representar essa relação, cada ponto representa o valor de uma variável no eixo horizontal e o valor de outra variável no eixo vertical.
